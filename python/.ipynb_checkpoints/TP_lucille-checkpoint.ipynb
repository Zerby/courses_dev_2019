```python
import csv
```

```python
only2000 = []
diplomas= []
geo = []
with open('37100035.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=';')
    for row in spamreader:
        if row[0] == "2000":
            only2000.append(row)
            diplomas.append(row[3])
            geo.append(row[1])
            
diplomas = set(diplomas)
maximum  = {}
for diploma in diplomas:
    dictMax=[]
    
    for row in only2000:
        if diploma == row[3]:
            if row[12] == "":row[12]=0
            dictMax.append(int(row[12]))

    maximum[diploma] = max(dictMax)

i=0
for key, value in sorted(maximum.items(), key=lambda item: item[1], reverse=True):
    print(key, value)
    i+=1
    if(i>2):break

 
```

```python
# Quels sont les trois diplomes ayant des personnes gagnant le plus ?
# Reponse -> Doctorat , Maitrise et Baccalauréat
```

```python
geo = set(geo)
for province in geo :
    print(province)
    for diploma in diplomas:
        dictMax = []
        for row in only2000:
            if diploma == row[3] and province == row[1]:
                if row[12] == "":row[12]=0
                dictMax.append(int(row[12]))
        dictMax = sum(dictMax) / len(dictMax)
        print("Moyenne:", dictMax, "$ pour un(e) ", diploma)
```
